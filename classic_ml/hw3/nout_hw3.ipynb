### Итоговое решение получилось: standart_scale+pca(8 компонент)+kmeans(8 кластеров)

```python
from scipy import sparse
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import StandardScaler

train = sparse.load_npz("train.npz")
```

```python
norm  = StandardScaler()
reducer = PCA(n_components = 8)
train_scaled = norm.fit_transform(train.toarray())
train_scaled_reduced = reducer.fit_transform(train_scaled)
kmeans = KMeans(n_clusters = 8)
kmeans.fit(train_scaled_reduced)
labels = kmeans.labels_
```

```python
import pandas as pd
test = pd.read_csv("sample_submission.csv")
```

```python
test["TARGET"] = labels
test.to_csv("solut.csv", index=False)
```
